Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.cluster import KMeans
```

Data

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/Datasets/Iris Flower/IRIS.csv")

#sb.pairplot(df,hue="species")                  #Data scatter view with labels         
sb.pairplot(df)                                 #Data scatter view without labels 

data = np.array(df.drop('species',axis = 1))    ##Removing the Labels Column 
```

Model

```python
kmeans = KMeans(n_clusters=3, random_state=0)   #K-Means: 3 Clusters
                           
```

Train


```python
kmeans.fit(data)                              #Training
```

Visualizao

```python
df['K-Means Classification'] = kmeans.labels_    #Inserindo uma coluna com os agrupamentos K-Means

df
```

```python
sb.pairplot(df,hue='K-Means Classification')     #View Data dispersion with K-Means labels
sb.pairplot(df,hue='species')                     #View Data dispersion with original labels
```
